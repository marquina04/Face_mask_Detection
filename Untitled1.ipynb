```python
import numpy as np
import os
import matplotlib.pyplot as plt
from imutils import paths
from tensorflow.keras.applications import MobileNetV2
# from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import classification_report
```

```python
from tensorflow.keras.layers import AveragePooling2D
```

```python
dataset=r'./Mask Dataset'
imagePaths= list(paths.list_images(dataset))
```

```python
imagePaths
```

```python
data=[]
labels=[]

for i in imagePaths:
    label=i.split(os.path.sep)[-2]
    labels.append(label)
    image=load_img(i,target_size=(224,224))
    image= img_to_array(image)
    image=preprocess_input(image)
    data.append(image)
```

```python
data=np.array(data,dtype='float32')
labels = np.array(labels)
```

```python
lb=LabelBinarizer()
labels =lb.fit_transform(labels)
labels = to_categorical(labels)
```

```python
print(len(data))
print(len(labels))
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(data,labels,stratify=labels,test_size=0.2,random_state=10)
```

```python
aug=ImageDataGenerator(rotation_range=40,zoom_range=0.5,width_shift_range=0.3, height_shift_range=0.3,shear_range=0.2,horizontal_flip=True, vertical_flip=True, fill_mode='nearest')
```

```python
baseModel= MobileNetV2(weights='imagenet',include_top=False, input_tensor=Input(shape=(224,224,3)))
```

```python
baseModel.summary()
```

```python
headModel= baseModel.output
headModel= AveragePooling2D(pool_size=(7,7))(headModel)
headModel= Flatten(name='Flatten')(headModel)
headModel= Dense(128, activation='relu')(headModel)
headModel=Dropout(0.5)(headModel)
headModel=Dense(2,activation='softmax')(headModel)

model=Model(inputs=baseModel.input, outputs=headModel)
```

```python
for layer in baseModel.layers:
    layer.trainable=False
    
```

```python
learning_rate=0.001
epoch=20
BS=12
opt=Adam(lr=0.001,decay=learning_rate/epoch)
model.compile(loss='binary_crossentropy',optimizer=opt, metrics=['accuracy'])

model.fit(aug.flow(X_train,Y_train,batch_size=12),steps_per_epoch=len(X_train)//12,validation_data=(X_test,Y_test), validation_steps=len(X_test)//12,epochs=20)
```

```python
# model.save(r'C:\Users\Ankit Nag\Dropbox\My PC (LAPTOP-MIS1LSEU)\Desktop\miscellaneous\Face mask detection\mobilenet_v2.model')
model.save('mobilenet_v2.h5')
```

```python
# model = r'./mobilenet_v2.model/'
predict= model.predict(X_test,batch_size=12)
predict=np.argmax(predict,axis=1)
print(classification_report(Y_test.argmax(axis=1), predict, target_names=lb.classes_))
```
